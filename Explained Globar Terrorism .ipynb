##### IMPORTING NECESSARY PACKAGES

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

**LOADING THE DATA**

```python
data = pd.read_csv(r"E:\globalterrorismdb_0718dist.csv", encoding='ISO-8859-1' )
data
```

```python
data.head()
```

```python
data.describe()
```

```python
data.columns
```

**CHECKING FOR NULL VALUES AND DROPPING THEM**

```python
data.isnull().sum()
```

```python
data.dropna(axis=1, how='all')
```

**TAKING ONLY IMPORTANT COLUMNS INTO DATAFRAME**

```python
data.rename(columns={'iyear':'Year', 'imonth':'Month', 'iday':'Day','attacktype1_txt':'Attacktype','target1' : 'Target','country_txt':'Country',
                 'region_txt':'Region', 'provstate': 'state','targtype1_txt':'Targtype','gname':'Group', 'claimed':'Claimed',
                 'weaptype1_txt':'Weaptype','nkill':'kill','nwound':'wounded','summary': 'Summary',},inplace=True)
```

```python
data = data[['Year','Month','Day','Country', 'Region', 'state', 'latitude','longitude','Attacktype','kill','wounded','Target',
            'Summary','Group','Targtype','Weaptype','motive' ]]
data['wounded']=data['wounded'].fillna(0).astype(int)
data['kill']=data['kill'].fillna(0).astype(int)
data['casualities']= data['kill'] + data['wounded']
```

```python
data.head()
```

**ANALYSING THE DATA THROUGH HISTOGRAM PLOTS**

```python
data.hist(figsize=(24,12))
```

**PLOTTING CORRELATION**

```python
plt.figure(figsize=(15,10))
sns.heatmap(np.round(data.corr(),2),annot=True, cmap='YlOrRd')
```

#### GENERAL ANALYSIS

```python
print('Country with the highest number of Terrorist Attacks:',data['Country'].value_counts().index[0])
print('Regions with the highest number of  Terrorist Attacks:',data['Region'].value_counts().index[0])
print('Maximum number of people were killed by a single terrorist attack are',data['kill'].max(),'people that took place in',data.loc[data['kill'].idxmax()].Country)
print('Year with the most Attacks:',data['Year'].value_counts().idxmax())
print('Month with the most Attacks:',data['Month'].value_counts().idxmax())
print('Group with the most Attacks:',data['Group'].value_counts().index[1])
print('Most Attacks Types:',data['Attacktype'].value_counts().idxmax())
```

#### THE BELOW PLOTS SHOWS THE NUMBER OF TERRORIST ACTIVITY TOOK PLACE EACH YEAR
**AS WE CAN SEE THAT IN THE YEAR 2014, IS THE YEAR WHEN MAXIMUM TERRORIST ATTACKS TOOK PLACE AND FOLLOWED BY 2015 AS THE SECOND HIGHEST**

```python
plt.subplots(figsize=(15,6))
sns.countplot(x= 'Year', data=data, palette='coolwarm',edgecolor=sns.color_palette('dark',7))
plt.xticks(rotation=90)
plt.title('Number of Terrorist Activity each Year')
plt.show()
```

**THE BAR PLOT BELOW SHOWS THE VARIOUS ATTACKING METHODS USED**
**SHOWING BOMBING/EXPLOSION IS THE MOST USED ATTACKING METHOD BY TERRORIST GROUP**

```python
plt.subplots(figsize=(15,6))
sns.countplot(x= 'Attacktype', data=data, palette='coolwarm',order=data['Attacktype'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Attacking Methods by Terrorists')
plt.show()
```

**THE BAR PLOT BELOW SHOWS THE MAJOR TERRORISTS TARGERS**

**MAXIMUM TARGETS WERE ON PRIVATE CITIZEN AND PROPERTY**

```python
plt.subplots(figsize=(15,6))
sns.countplot(x= 'Targtype', data=data, palette='flare',order=data['Targtype'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Terrorists Targets')
plt.show()
```

**THE BAR PLOT BELOW SHOWS MOST AFFECTED TERRORIST TARGETS REGION WISE**

**SHOWS THAT MIDDLE EAST AND NORTH AFRICA IS THE MOST TARGETED AND AFFECTED AREA**

```python
sns.countplot(x= 'Region', data=data, palette='Set2',edgecolor=sns.color_palette('dark',7),order=data['Region'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Terrorists Targets by Region')
plt.show()
```

**THE BAR DIAGRAM BELOW SHOWS THE MOST AFFECTED COUNTRIES**

```python
plt.subplots(figsize=(20,6))
sns.barplot(data['Country'].value_counts()[:15].index,data['Country'].value_counts()[:15].values,palette="YlOrBr")
plt.title('Most Affected Countries')
plt.show()
```

**THIS PLOT SHOWS THE NUMBER OF ATTACK TOOK PLACE AND NO OF PEOPLE GOT KILLED COUNTRY WISE**

```python
count_terror = data['Country'].value_counts()[:15].to_frame()
count_terror.columns=['Attacks']
count_kill=data.groupby ('Country')['kill'].sum().to_frame()
count_terror.merge(count_kill,left_index = True,right_index =True,how='left').plot.bar(width=0.9)
fig=plt.gcf()
fig.set_size_inches(16,4)
plt.show()
```
